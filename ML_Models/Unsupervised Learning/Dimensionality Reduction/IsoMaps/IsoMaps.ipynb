# **IsoMaps Model Theory**


## Isomap (Isometric Mapping)

---

## Theory
Isomap (Isometric Mapping) is a non-linear dimensionality reduction technique that extends the concept of Multidimensional Scaling (MDS) by preserving the geodesic distances between data points rather than Euclidean distances. It is particularly useful for datasets where the intrinsic geometry is non-linear, such as manifolds. Isomap is widely used in data visualization, pattern recognition, and exploratory data analysis.

The main idea is to:
- Construct a neighborhood graph to approximate the manifold structure.
- Compute the geodesic distances (shortest paths) between all pairs of points.
- Apply Multidimensional Scaling (MDS) to the geodesic distances to obtain a low-dimensional embedding.

---

## Mathematical Foundation
- **Neighborhood Graph**:
  Construct a graph where each data point is connected to its \( k \) nearest neighbors or all points within a radius \( \epsilon \).

- **Geodesic Distance**:
  The geodesic distance \( d_G(i, j) \) between two points \( i \) and \( j \) is the shortest path distance on the neighborhood graph. It is computed using algorithms like Dijkstra's or Floyd-Warshall:
  $$ d_G(i, j) = \text{shortest path distance between } i \text{ and } j $$

- **Multidimensional Scaling (MDS)**:
  Apply MDS to the geodesic distance matrix \( D_G \) to find a low-dimensional embedding \( Y \) that preserves the distances:
  $$ Y = \arg\min_Y \sum_{i < j} (d_G(i, j) - \|y_i - y_j\|)^2 $$

---

## Algorithm Steps
1. **Construct Neighborhood Graph**:
   - For each data point, connect it to its \( k \) nearest neighbors or all points within a radius \( \epsilon \).

2. **Compute Geodesic Distances**:
   - Compute the shortest path distances between all pairs of points on the neighborhood graph.

3. **Apply Multidimensional Scaling (MDS)**:
   - Apply MDS to the geodesic distance matrix to obtain the low-dimensional embedding.

---

## Key Parameters
- **n_neighbors**: The number of neighbors to consider for each point.
- **n_components**: The number of dimensions in the low-dimensional embedding.
- **path_method**: The algorithm to use for computing shortest paths (e.g., `auto`, `Dijkstra`, `Floyd-Warshall`).

---

## Advantages
- Preserves the intrinsic geometry of the data.
- Effective for non-linear manifolds.
- Provides a low-dimensional representation that captures the global structure.
- Works well for data visualization.

---

## Disadvantages
- Computationally expensive for large datasets (\( O(n^3) \) for geodesic distance computation).
- Sensitive to the choice of neighborhood size \( k \).
- Struggles with noisy data.
- Requires careful tuning of parameters.

---

## Implementation Tips
- Use **k-nearest neighbors** to construct the neighborhood graph.
- Choose an appropriate **n_neighbors** value to balance local and global structure.
- Scale features before applying Isomap to ensure equal contribution from all features.
- Consider **Landmark Isomap** for large datasets to reduce computational cost.

---

## Applications
- Data visualization (e.g., 2D/3D plots of high-dimensional data)
- Pattern recognition
- Exploratory data analysis
- Image processing
- Bioinformatics

Isomap is a powerful and flexible technique for non-linear dimensionality reduction. While it requires careful parameter tuning and can be computationally expensive, it is a valuable tool for many real-world applications.

# Model Evaluation for Isometric Mapping (Isomap)

---

### 1. Reconstruction Error
**Formula:**
$$
\text{Reconstruction Error} = \frac{1}{N} \sum_{i=1}^N ||x_i - \hat{x}_i||^2
$$
**Description:**
- Measures the difference between the original data points and the reconstructed data points in the low-dimensional space.
- Useful for assessing the quality of the embedding.

**Interpretation:**
- Lower values indicate better reconstruction.
- Useful for evaluating the effectiveness of the Isomap.

---

### 2. Mean Squared Error (MSE)
**Formula:**
$$
\text{MSE} = \frac{1}{N} \sum_{i=1}^N (x_i - \hat{x}_i)^2
$$
**Description:**
- Measures the average squared difference between the original and reconstructed data points.

**Interpretation:**
- Lower values indicate better reconstruction.
- Commonly used for continuous data.

---

### 3. Trustworthiness
**Formula:**
$$
\text{Trustworthiness} = 1 - \frac{2}{n(n-1)} \sum_{i=1}^n \sum_{j \in N_k(x_i)} \mathbb{I}(y_j \notin N_k(y_i))
$$
**Description:**
- Measures the degree to which the nearest neighbors in the high-dimensional space are preserved in the low-dimensional space.

**Interpretation:**
- Higher values indicate better preservation of local structure.
- Useful for evaluating the reliability of the embedding.

---

### 4. Continuity
**Formula:**
$$
\text{Continuity} = 1 - \frac{2}{n(n-1)} \sum_{i=1}^n \sum_{j \in N_k(y_i)} \mathbb{I}(x_j \notin N_k(x_i))
$$
**Description:**
- Measures the degree to which the nearest neighbors in the low-dimensional space are preserved from the high-dimensional space.

**Interpretation:**
- Higher values indicate better preservation of local structure.
- Complementary to trustworthiness.

---

### 5. Stress
**Formula:**
$$
\text{Stress} = \sqrt{\frac{\sum_{i < j} (d_{ij} - \tilde{d}_{ij})^2}{\sum_{i < j} d_{ij}^2}}
$$
**Description:**
- Measures the overall discrepancy between the distances in the high-dimensional and low-dimensional spaces.

**Interpretation:**
- Lower values indicate better preservation of distances.
- Useful for assessing the quality of the embedding.

---

### 6. Silhouette Score
**Formula:**
$$
\text{Silhouette Score} = \frac{b - a}{\max(a, b)}
$$
**Description:**
- Measures how similar an object is to its own cluster compared to other clusters.
- Useful for evaluating the clustering quality of the Isomap embedding.

**Interpretation:**
- Values range from -1 to 1.
- Higher values indicate better clustering.

---

### 7. Visualization Quality
**Description:**
- Evaluates the visual interpretability of the Isomap embedding.
- Subjective measure based on the clarity of clusters and separation of different classes.

**Interpretation:**
- Clear and well-separated clusters indicate high-quality visualization.
- Useful for exploratory data analysis.

---

### 8. Dimensionality Reduction Effectiveness
**Description:**
- Evaluates the effectiveness of Isomap in reducing the number of dimensions while retaining information.
- Measured by comparing performance metrics (e.g., classification accuracy) before and after dimensionality reduction.

**Interpretation:**
- Higher retained performance indicates effective dimensionality reduction.

---

### 9. Computational Efficiency
**Description:**
- Evaluates the computational cost of performing Isomap, including time and memory usage.
- Important for large datasets or real-time applications.

**Interpretation:**
- Lower computational cost indicates better scalability.
- Useful for assessing the practicality of Isomap for specific use cases.

---

### 10. Sensitivity to Parameters
**Description:**
- Evaluates how sensitive the Isomap results are to changes in parameters such as the number of neighbors \( k \) and the number of dimensions \( d \).
- Important for ensuring robust and consistent results.

**Interpretation:**

---

### 11. Outlier Detection
**Description:**

**Interpretation:**

### 12. Overfitting and Underfitting
**Description:**

# IsoMaps - Example

## Data loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.manifold import Isomap
from sklearn.preprocessing import StandardScaler

# Data Loading
# Load the Digits dataset from sklearn.datasets
data = load_digits()
X = data.data
y = data.target
```

##  Data processing

```python
# Data Processing
# Convert the data to a DataFrame for better visualization
df = pd.DataFrame(data.data, columns=[f'pixel{i}' for i in range(64)])
df['target'] = data.target

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Plotting data

```python
# Data Plotting
# Plot a few sample images from the dataset
n_samples_to_plot = 5
fig, axes = plt.subplots(1, n_samples_to_plot, figsize=(15, 3))
for i in range(n_samples_to_plot):
    axes[i].imshow(X_scaled[i].reshape(8, 8), cmap='gray')
    axes[i].set_title(f"Label: {y[i]}")
    axes[i].axis('off')
plt.show()
```

## Model definition

```python

# Model Definition
# Define the Isomap model
n_components = 2  # Number of components to extract
n_neighbors = 5   # Number of neighbors to consider for the neighborhood graph
isomap = Isomap(n_components=n_components, n_neighbors=n_neighbors)

# Model Training
# Fit the Isomap model to the scaled data
X_isomap = isomap.fit_transform(X_scaled)

# Convert the transformed data to a DataFrame
df_isomap = pd.DataFrame(X_isomap, columns=[f'Isomap{i+1}' for i in range(n_components)])
df_isomap['target'] = y
```

## Model evaulation

```python
# Data Plotting
# Plot the transformed data to visualize the Isomap components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Isomap1', y='Isomap2', hue='target', data=df_isomap, palette='viridis')
plt.title('Isomap Components (Isomap1 vs Isomap2)')
plt.show()

# Additional Analysis
# Print the reconstruction error
print("\nReconstruction Error:")
print(isomap.reconstruction_error_)

# Print the number of neighbors used
print("\nNumber of Neighbors Used:")
print(isomap.n_neighbors)

# Print the number of components
print("\nNumber of Components:")
print(isomap.n_components_)
```
