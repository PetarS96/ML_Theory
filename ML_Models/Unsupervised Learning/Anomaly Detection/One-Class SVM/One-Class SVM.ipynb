# **One-CLass SVM Model Theory**


## One-Class SVM

---

## Theory
One-Class SVM (Support Vector Machine) is a type of SVM used for anomaly detection and novelty detection. Unlike traditional SVMs, which are used for classification and regression tasks, One-Class SVM is designed to learn from a single class of data and identify outliers or novelties that do not belong to this class. It is particularly useful in scenarios where the data is imbalanced, and the normal class is well-represented, but the anomalies are rare or unknown.

The main idea is to:
- Find a hyperplane that best separates the data from the origin in a high-dimensional feature space.
- Maximize the margin between the data and the hyperplane to capture the essence of the normal class.
- Identify data points that fall outside this margin as anomalies or novelties.

---

## Mathematical Foundation
- **Feature Space Mapping**:
  The data points \( x_i \) are mapped to a high-dimensional feature space using a kernel function \( \phi \):
  \[ \phi(x_i) \]

- **Hyperplane and Margin**:
  The goal is to find a hyperplane \( w \cdot \phi(x) + b = 0 \) such that:
  \[ \frac{1}{n} \sum_{i=1}^n \xi_i + \frac{1}{2} \| w \|^2 \]
  is minimized, subject to:
  \[ w \cdot \phi(x_i) + b \geq 1 - \xi_i \]
  \[ \xi_i \geq 0 \]
  - \( w \): Weight vector.
  - \( b \): Bias term.
  - \( \xi_i \): Slack variables to allow some data points to be within the margin.
  - \( \nu \): A parameter that controls the trade-off between the number of outliers and the margin size.

- **Kernel Trick**:
  The kernel trick is used to compute the dot product in the high-dimensional feature space without explicitly mapping the data:
  \[ K(x_i, x_j) = \phi(x_i) \cdot \phi(x_j) \]
  Common kernel functions include the linear kernel, polynomial kernel, and radial basis function (RBF) kernel.

---

## Algorithm Steps
1. **Data Preparation**:
   - Collect and preprocess the data, ensuring that it represents the normal class.

2. **Kernel Selection**:
   - Choose an appropriate kernel function (e.g., RBF, linear, polynomial).

3. **Model Training**:
   - Solve the optimization problem to find the weight vector \( w \) and bias term \( b \):
     \[ \min_{w, b, \xi} \frac{1}{2} \| w \|^2 + \frac{1}{\nu n} \sum_{i=1}^n \xi_i \]
     subject to:
     \[ w \cdot \phi(x_i) + b \geq 1 - \xi_i \]
     \[ \xi_i \geq 0 \]

4. **Compute Decision Function**:
   - Compute the decision function for each data point:
     \[ f(x) = w \cdot \phi(x) + b \]
   - Data points with \( f(x) < 0 \) are considered outliers or novelties.

5. **Identify Outliers**:
   - Identify data points that fall outside the margin as anomalies or novelties.

---

## Key Parameters
- **nu**: A parameter between 0 and 1 that controls the trade-off between the number of outliers and the margin size. It is an upper bound on the fraction of training points that can be considered outliers.
- **kernel**: The kernel function used to map the data into a high-dimensional feature space (e.g., `linear`, `poly`, `rbf`, `sigmoid`).
- **gamma**: The kernel coefficient for `rbf`, `poly`, and `sigmoid` kernels. It can be specified as 'scale' (default) or 'auto', or a custom value.
- **degree**: The degree of the polynomial kernel function (only used for `poly` kernel).
- **coef0**: The independent term in the kernel function (used for `poly` and `sigmoid` kernels).
- **tol**: Tolerance for stopping criterion.
- **shrinking**: Whether to use the shrinking heuristic.
- **cache_size**: The size of the kernel cache (in MB).
- **max_iter**: The maximum number of iterations for the solver.

---

## Advantages
- **Anomaly Detection**: Effective for identifying outliers and novelties in datasets where the normal class is well-represented.
- **Non-linear Boundaries**: Can handle non-linear boundaries using kernel functions.
- **Robustness**: Robust to noise and outliers in the training data.
- **Parameter Control**: The `nu` parameter provides control over the trade-off between the number of outliers and the margin size.

---

## Disadvantages
- **Computational Cost**: Can be computationally expensive, especially for large datasets and high-dimensional data.
- **Parameter Sensitivity**: Sensitive to the choice of the `nu` parameter and the kernel parameters.
- **Interpretability**: The decision boundary and the choice of parameters can be difficult to interpret.
- **Training Time**: Training can be time-consuming, especially for large datasets.

---

## Implementation Tips
- **Choose `nu` Carefully**: The `nu` parameter should be chosen based on the expected fraction of outliers in the data. A smaller `nu` value will result in a larger margin and fewer outliers.
- **Select Appropriate Kernel**: Choose the kernel function based on the nature of the data. The RBF kernel is a good default choice for non-linear data.
- **Tune Kernel Parameters**: For the RBF kernel, the `gamma` parameter is crucial. Use cross-validation to find the optimal value.
- **Preprocess the Data**: Standardize or normalize the data to improve the performance of the kernel functions.
- **Monitor Training Time**: Be aware of the computational cost, especially for large datasets. Consider using a subset of the data for initial experimentation.
- **Evaluate Performance**: Use metrics like precision, recall, and F1-score to evaluate the performance of the One-Class SVM on a validation set.

---

## Applications
- **Anomaly Detection**: Identifying outliers in datasets where the normal class is well-represented.
- **Novelty Detection**: Detecting new or unknown patterns in data.
- **Fraud Detection**: Identifying fraudulent transactions or activities.
- **Network Intrusion Detection**: Detecting unusual network traffic patterns.
- **Medical Diagnostics**: Identifying abnormal medical conditions or anomalies in medical data.
- **Quality Control**: Detecting defects or anomalies in manufacturing processes.

One-Class SVM is a powerful tool for anomaly and novelty detection. While it has some limitations, it is particularly useful in scenarios where the normal class is well-represented, and the anomalies are rare or unknown.

## Model Evaluation for One-Class SVM

---

### 1. Accuracy Score
**Formula:**
$$
\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
$$
**Description:**
- Measures the overall prediction accuracy of the One-Class SVM.
- Suitable for evaluating the model's ability to correctly identify inliers and outliers.

**Interpretation:**
- Higher values indicate better overall performance.
- Should be used alongside other metrics for imbalanced datasets.

---

### 2. Precision
**Formula:**
$$
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
$$
**Description:**
- Measures the proportion of correctly predicted inliers out of all predicted inliers.
- Useful when the cost of false positives (outliers misclassified as inliers) is high.

**Interpretation:**
- Higher values indicate fewer false positives.
- Should be balanced with recall for imbalanced datasets.

---

### 3. Recall (Sensitivity)
**Formula:**
$$
\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$
**Description:**
- Measures the proportion of correctly predicted inliers out of all actual inliers.
- Useful when the cost of false negatives (inliers misclassified as outliers) is high.

**Interpretation:**
- Higher values indicate fewer false negatives.
- Should be balanced with precision for imbalanced datasets.

---

### 4. F1 Score
**Formula:**
$$
\text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$
**Description:**
- Harmonic mean of precision and recall.
- Useful for imbalanced datasets where both false positives and false negatives matter.

**Interpretation:**
- Higher values indicate better balance between precision and recall.
- Maximum value is 1.0 (perfect precision and recall).

---

### 5. ROC-AUC Score
**Formula:**
$$
\text{AUC} = \int_{0}^{1} \text{TPR}(t) \cdot \text{FPR}'(t) \, dt
$$
**Description:**
- Area under the ROC curve for binary classification.
- Threshold-independent performance measure.

**Interpretation:**
- 1.0 indicates perfect classification.
- 0.5 represents random prediction.
- Useful for evaluating the model's ability to distinguish inliers from outliers.

---

### 6. Confusion Matrix
**Description:**
- A table showing the counts of true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN).
- Provides a detailed breakdown of classification performance.

**Interpretation:**
- Diagonal elements (TP, TN) represent correct predictions.
- Off-diagonal elements (FP, FN) represent errors.
- Useful for understanding the model's performance in detail.

---

### 7. Outlier Detection Rate
**Description:**
- Measures the proportion of true outliers correctly identified by the model.
- Evaluates the model's ability to detect anomalies.

**Interpretation:**
- Higher values indicate better outlier detection.
- Useful for assessing the model's effectiveness in anomaly detection tasks.

---

### 8. False Positive Rate (FPR)
**Formula:**
$$
\text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}
$$
**Description:**
- Measures the proportion of inliers incorrectly classified as outliers.
- Evaluates the model's tendency to generate false alarms.

**Interpretation:**
- Lower values indicate fewer false positives.
- Useful for applications where false alarms are costly.

---

### 9. Decision Function Values
**Description:**
- Represents the signed distance of samples to the separating hyperplane.
- Positive values indicate inliers, while negative values indicate outliers.

**Interpretation:**
- Higher absolute values indicate stronger confidence in the prediction.
- Useful for understanding the model's decision boundaries.

---

### 10. Parameter Sensitivity Analysis
**Description:**
- Evaluates the impact of One-Class SVM parameters (e.g., `nu` and `kernel`) on model performance.
- Involves testing different parameter combinations and measuring performance metrics.

**Interpretation:**
- Helps identify optimal parameter settings for the dataset.
- Critical for achieving meaningful results.

---

## Data loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report

# Data Loading
# Load the Iris dataset from sklearn.datasets
data = load_iris()
X = data.data
y = data.target

# For simplicity, we'll use only one class (e.g., class 0) and treat the rest as outliers
X_class_0 = X[y == 0]
y_class_0 = y[y == 0]

# Data Processing
# Convert the data to a DataFrame for better visualization
df = pd.DataFrame(X_class_0, columns=data.feature_names)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

##  Data processing

```python

# Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_class_0)

# Model Definition
# Define the One-Class SVM model
nu = 0.1  # Anomaly detection parameter (0 < nu < 1)
kernel = 'rbf'  # Kernel type
gamma = 'scale'  # Kernel coefficient
one_class_svm = OneClassSVM(nu=nu, kernel=kernel, gamma=gamma)

# Model Training
# Fit the One-Class SVM model to the data
one_class_svm.fit(X_scaled)
```

## Plotting data

## Model definition

```python

# Model Evaluation
# Predict the labels (1 for inliers, -1 for outliers)
y_pred = one_class_svm.predict(X_scaled)

# Convert predictions to binary labels (0 for inliers, 1 for outliers)
y_pred = np.where(y_pred == 1, 0, 1)
```

## Model evaulation

```python

# Print the confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_class_0, y_pred))

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_class_0, y_pred))

# Data Plotting
# Plot the data points and the decision boundary
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='viridis')
plt.title('One-Class SVM (First Two Features)')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])

# Plot the decision boundary
xx, yy = np.meshgrid(np.linspace(X_scaled[:, 0].min(), X_scaled[:, 0].max(), 500),
                     np.linspace(X_scaled[:, 1].min(), X_scaled[:, 1].max(), 500))
Z = one_class_svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='red')
plt.show()
```
