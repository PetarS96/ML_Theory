# **Multidimensional Scaling Model Theory**


## Multidimensional Scaling (MDS)

---

## Theory
Multidimensional Scaling (MDS) is a dimensionality reduction technique used to visualize the similarity or dissimilarity between data points in a low-dimensional space. It aims to preserve the pairwise distances between points as much as possible. MDS is widely used in data visualization, psychology, market research, and bioinformatics.

The main idea is to:
- Start with a distance matrix representing the pairwise distances between data points.
- Find a low-dimensional embedding where the distances between points approximate the original distances.

---

## Mathematical Foundation
- **Distance Matrix**:
  The distance matrix \( D \) contains the pairwise distances between data points:
  $$ D_{ij} = d(x_i, x_j) $$
  - \( d(x_i, x_j) \): Distance between points \( x_i \) and \( x_j \) (e.g., Euclidean distance).

- **Objective**:
  Find a low-dimensional embedding \( Y \) that minimizes the stress function:
  $$ \text{Stress} = \sqrt{\frac{\sum_{i < j} (d(x_i, x_j) - \|y_i - y_j\|)^2}{\sum_{i < j} d(x_i, x_j)^2}} $$
  - \( y_i, y_j \): Low-dimensional representations of \( x_i \) and \( x_j \).

- **Classical MDS**:
  If the distance matrix is Euclidean, classical MDS can be solved using eigenvalue decomposition:
  1. Compute the Gram matrix \( B = -\frac{1}{2} H D^2 H \), where \( H \) is the centering matrix.
  2. Perform eigenvalue decomposition on \( B \) to obtain eigenvalues \( \lambda_i \) and eigenvectors \( v_i \).
  3. The low-dimensional embedding is given by:
     $$ Y = V_k \Lambda_k^{1/2} $$
     - \( V_k \): Matrix of the top \( k \) eigenvectors.
     - \( \Lambda_k \): Diagonal matrix of the top \( k \) eigenvalues.

---

## Algorithm Steps
1. **Compute Distance Matrix**:
   - Compute the pairwise distance matrix \( D \) from the input data.

2. **Apply MDS**:
   - For classical MDS:
     1. Compute the Gram matrix \( B \).
     2. Perform eigenvalue decomposition on \( B \).
     3. Obtain the low-dimensional embedding \( Y \).
   - For metric or non-metric MDS, use optimization techniques to minimize the stress function.

---

## Key Parameters
- **n_components**: The number of dimensions in the low-dimensional embedding.
- **metric**: Whether to use metric MDS (preserve distances) or non-metric MDS (preserve rank order of distances).
- **dissimilarity**: The type of dissimilarity measure to use (e.g., `euclidean`, `precomputed`).

---

## Advantages
- Preserves pairwise distances or dissimilarities.
- Effective for data visualization.
- Works well for both metric and non-metric data.
- Provides interpretable low-dimensional representations.

---

## Disadvantages
- Computationally expensive for large datasets (\( O(n^3) \) for classical MDS).
- Sensitive to noise and outliers.
- Struggles with high-dimensional data.
- Requires careful tuning of parameters.

---

## Implementation Tips
- Use **precomputed distances** if the distance matrix is already available.
- Choose an appropriate **n_components** value for visualization (e.g., 2 or 3).
- Scale features before applying MDS to ensure equal contribution from all features.
- Consider **Sammon Mapping** or **t-SNE** for non-linear dimensionality reduction.

---

## Applications
- Data visualization (e.g., 2D/3D plots of high-dimensional data)
- Psychology (e.g., perception studies)
- Market research (e.g., customer preference analysis)
- Bioinformatics (e.g., gene expression analysis)
- Social network analysis

Multidimensional Scaling is a powerful and flexible technique for dimensionality reduction and visualization. While it has limitations, it is a valuable tool for many real-world applications.

# Model Evaluation for Multidimensional Scaling (MDS)

---

### 1. Stress
**Formula:**
$$
\text{Stress} = \sqrt{\frac{\sum_{i < j} (d_{ij} - \tilde{d}_{ij})^2}{\sum_{i < j} d_{ij}^2}}
$$
**Description:**
- Measures the overall discrepancy between the distances in the high-dimensional and low-dimensional spaces.
- Commonly used loss function in MDS.

**Interpretation:**
- Lower values indicate better preservation of distances.
- Useful for assessing the quality of the embedding.

---

### 2. Kruskal's Stress-1
**Formula:**
$$
\text{Stress-1} = \sqrt{\frac{\sum_{i < j} (d_{ij} - \tilde{d}_{ij})^2}{\sum_{i < j} \tilde{d}_{ij}^2}}
$$
**Description:**
- Similar to stress, but normalized by the sum of the squared distances in the low-dimensional space.

**Interpretation:**
- Lower values indicate better preservation of distances.
- Useful for comparing different MDS embeddings.

---

### 3. Sammon's Stress
**Formula:**
$$
\text{Sammon's Stress} = \frac{1}{\sum_{i < j} d_{ij}} \sum_{i < j} \frac{(d_{ij} - \tilde{d}_{ij})^2}{d_{ij}}
$$
**Description:**
- Measures the weighted discrepancy between the distances in the high-dimensional and low-dimensional spaces.
- Weights the errors by the inverse of the original distances.

**Interpretation:**
- Lower values indicate better preservation of distances.
- Useful for emphasizing the preservation of small distances.

---

### 4. Trustworthiness
**Formula:**
$$
\text{Trustworthiness} = 1 - \frac{2}{n(n-1)} \sum_{i=1}^n \sum_{j \in N_k(x_i)} \mathbb{I}(y_j \notin N_k(y_i))
$$
**Description:**
- Measures the degree to which the nearest neighbors in the high-dimensional space are preserved in the low-dimensional space.

**Interpretation:**
- Higher values indicate better preservation of local structure.
- Useful for evaluating the reliability of the embedding.

---

### 5. Continuity
**Formula:**
$$
\text{Continuity} = 1 - \frac{2}{n(n-1)} \sum_{i=1}^n \sum_{j \in N_k(y_i)} \mathbb{I}(x_j \notin N_k(x_i))
$$
**Description:**
- Measures the degree to which the nearest neighbors in the low-dimensional space are preserved from the high-dimensional space.

**Interpretation:**
- Higher values indicate better preservation of local structure.
- Complementary to trustworthiness.

---

### 6. Silhouette Score
**Formula:**
$$
\text{Silhouette Score} = \frac{b - a}{\max(a, b)}
$$
**Description:**
- Measures how similar an object is to its own cluster compared to other clusters.
- Useful for evaluating the clustering quality of the MDS embedding.

**Interpretation:**
- Values range from -1 to 1.
- Higher values indicate better clustering.

---

### 7. Visualization Quality
**Description:**
- Evaluates the visual interpretability of the MDS embedding.
- Subjective measure based on the clarity of clusters and separation of different classes.

**Interpretation:**
- Clear and well-separated clusters indicate high-quality visualization.
- Useful for exploratory data analysis.

---

### 8. Dimensionality Reduction Effectiveness
**Description:**
- Evaluates the effectiveness of MDS in reducing the number of dimensions while retaining information.
- Measured by comparing performance metrics (e.g., classification accuracy) before and after dimensionality reduction.

**Interpretation:**
- Higher retained performance indicates effective dimensionality reduction.
- Useful for assessing the trade-off between complexity and information loss.

---

### 9. Computational Efficiency
**Description:**
- Evaluates the computational cost of performing MDS, including time and memory usage.
- Important for large datasets or real-time applications.

**Interpretation:**
- Lower computational cost indicates better scalability.
- Useful for assessing the practicality of MDS for specific use cases.

---

### 10. Sensitivity to Parameters
**Description:**
- Evaluates how sensitive the MDS results are to changes in parameters such as the number of dimensions \( d \) and the type of distance metric used.
- Important for ensuring robust and consistent results.

**Interpretation:**
- Results that are stable across different parameter settings are more reliable.
- Useful for tuning and validating the model.

---

### 11. Outlier Detection
**Description:**
- Evaluates the ability of MDS to detect outliers by examining the distance of data points from the cluster centers in the low-dimensional space.

**Interpretation:**
- Points far from cluster centers are potential outliers.
- Useful for identifying anomalies in the data.

---

### 12. Overfitting and Underfitting
**Description:**
- Evaluates the MDS's performance on training and validation sets.
- Overfitting occurs when the model performs well on the training set but poorly on the validation set.
- Underfitting occurs when the model performs poorly on both sets.

**Interpretation:**
- Balanced performance on both sets indicates a well-trained model.
- Useful for tuning hyperparameters and model architecture.

---

### 13. Learning Curves
**Description:**
- Plots the training and validation loss over iterations.
- Helps identify overfitting, underfitting, and optimal training duration.

**Interpretation:**
- Convergence of training and validation loss indicates a well-trained model.
- Useful for monitoring training progress and adjusting hyperparameters.

---

### 14. Factor Interpretability
**Description:**
- Evaluates the interpretability of the low-dimensional factors by examining the embedded coordinates.
- Useful for understanding the meaning of the factors in the context of the data.

**Interpretation:**
- Factors with clear and meaningful patterns indicate good interpretability.
- Useful for domain-specific analysis.

---

### 15. Model Fit Indices
**Description:**
- Evaluates the fit of the MDS model using various indices such as the Akaike Information Criterion (AIC) and the Bayesian Information Criterion (BIC).
- Useful for assessing the overall fit of the model.

**Interpretation:**
- Lower AIC and BIC values indicate a better fit.
- Useful for model validation and comparison.

| verbose                    | Level of verbosity.                                                                                                                                | 0                |

# Multidimensional Scaling - Example

## Data loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
from sklearn.preprocessing import StandardScaler

# Data Loading
# Load the Digits dataset from sklearn.datasets
data = load_digits()
X = data.data
y = data.target

# Data Processing
# Convert the data to a DataFrame for better visualization
df = pd.DataFrame(data.data, columns=[f'pixel{i}' for i in range(64)])
df['target'] = data.target

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

##  Data processing

```python

# Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Data Plotting
# Plot a few sample images from the dataset
n_samples_to_plot = 5
fig, axes = plt.subplots(1, n_samples_to_plot, figsize=(15, 3))
for i in range(n_samples_to_plot):
    axes[i].imshow(X_scaled[i].reshape(8, 8), cmap='gray')
    axes[i].set_title(f'Label: {y[i]}')
    axes[i].axis('off')
plt.show()
```

## Plotting data

## Model definition

```python

# Model Definition
# Define the MDS model
n_components = 2  # Number of components to extract
mds = MDS(n_components=n_components, random_state=42)

# Model Training
# Fit the MDS model to the scaled data
X_mds = mds.fit_transform(X_scaled)

# Convert the transformed data to a DataFrame
df_mds = pd.DataFrame(X_mds, columns=[f'MDS{i+1}' for i in range(n_components)])
df_mds['target'] = y
```

## Model evaulation

```python

# Data Plotting
# Plot the transformed data to visualize the MDS components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MDS1', y='MDS2', hue='target', data=df_mds, palette='viridis')
plt.title('MDS Components (MDS1 vs MDS2)')
plt.show()

# Additional Analysis
# Print the stress of the MDS model
print("\nStress of the MDS model:")
print(mds.stress_)

# Print the number of components
print("\nNumber of Components:")
print(mds.n_components)
```
