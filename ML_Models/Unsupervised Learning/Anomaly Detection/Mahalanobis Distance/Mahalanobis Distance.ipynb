# **Mahalanobis Distance Model Theory**


## Mahalanobis Distance

---

## Theory
Mahalanobis Distance is a distance metric that measures the distance between a point and a distribution. Unlike Euclidean distance, which treats all directions equally, Mahalanobis Distance accounts for correlations between variables and the shape of the data distribution. It is widely used in anomaly detection, classification, and clustering.

The main idea is to:
- Measure how far a data point is from the mean of the distribution.
- Normalize distances based on the covariance structure of the data.
- Identify data points that deviate significantly from the distribution as anomalies.

---

## Mathematical Foundation
- **Mean Vector**:
  The dataset is represented by \( X = \{ x_1, x_2, ..., x_n \} \), and the mean vector is:
  \[
  \mu = \frac{1}{n} \sum_{i=1}^{n} x_i
  \]

- **Covariance Matrix**:
  The covariance matrix \( S \) captures the relationships between variables:
  \[
  S = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)(x_i - \mu)^T
  \]

- **Mahalanobis Distance Formula**:
  The Mahalanobis Distance \( D_M(x) \) of a point \( x \) from the mean \( \mu \) is defined as:
  \[
  D_M(x) = \sqrt{(x - \mu)^T S^{-1} (x - \mu)}
  \]
  - \( x \): Data point.
  - \( \mu \): Mean vector.
  - \( S^{-1} \): Inverse of the covariance matrix.

- **Outlier Detection**:
  - Points with high Mahalanobis Distance values are considered anomalies.
  - A threshold based on the Chi-Square distribution can be used to detect anomalies.

---

## Algorithm Steps
1. **Compute Mean Vector**:
   - Calculate the mean \( \mu \) of the dataset.

2. **Compute Covariance Matrix**:
   - Compute the covariance matrix \( S \).

3. **Compute Mahalanobis Distance**:
   - For each data point \( x \), compute \( D_M(x) \) using:
     \[
     D_M(x) = \sqrt{(x - \mu)^T S^{-1} (x - \mu)}
     \]

4. **Set Threshold**:
   - Use a Chi-Square distribution with \( p \) degrees of freedom to determine a threshold.
   - If \( D_M(x) > \text{threshold} \), classify \( x \) as an anomaly.

---

## Key Parameters
- **Mean Vector \( \mu \)**: The central location of the dataset.
- **Covariance Matrix \( S \)**: Measures relationships between variables.
- **Threshold**: Defines the boundary between normal and anomalous points. Common choices:
  - \( \chi^2_{p, \alpha} \): Chi-Square value at significance level \( \alpha \).
- **Dimensionality \( p \)**: The number of features in the dataset.

---

## Advantages
- **Accounts for Correlations**: Considers relationships between variables.
- **Scale-Invariant**: Adjusts for different scales of data.
- **Effective for Multivariate Anomalies**: Detects anomalies in high-dimensional spaces.
- **Statistical Foundation**: Based on probability distributions.

---

## Disadvantages
- **Computational Cost**: Inverting the covariance matrix is expensive for large datasets.
- **Assumes Normal Distribution**: May not work well with non-Gaussian data.
- **Sensitive to Small Sample Sizes**: Requires enough data to estimate covariance accurately.
- **Curse of Dimensionality**: High-dimensional datasets can lead to numerical instability.

---

## Implementation Tips
- **Check Data Distribution**: Works best when data follows a multivariate normal distribution.
- **Regularize Covariance Matrix**: If the dataset is small or ill-conditioned, use a regularized covariance matrix.
- **Use Robust Estimators**: Consider robust covariance estimation methods (e.g., Minimum Covariance Determinant).
- **Choose an Appropriate Threshold**: Use domain knowledge or statistical tests to set the anomaly threshold.
- **Normalize Features**: Standardize data to avoid numerical issues in covariance calculations.
- **Evaluate Performance**: Use precision-recall metrics for anomaly detection evaluation.

---

## Applications
- **Anomaly Detection**: Identifying outliers in high-dimensional data.
- **Fraud Detection**: Detecting fraudulent activities in financial transactions.
- **Medical Diagnostics**: Identifying abnormal patterns in medical data.
- **Quality Control**: Detecting defective products in manufacturing.
- **Network Security**: Identifying unusual behavior in network traffic.
- **Biometric Authentication**: Detecting spoofed or fraudulent biometric data.

Mahalanobis Distance is a powerful tool for identifying outliers in multivariate datasets. While it has limitations, proper preprocessing and parameter tuning can make it highly effective for anomaly detection and classification tasks.


## **Model Evaluation for Mahalanobis Distance**

### 1. Mahalanobis Distance Formula

**Formula:**
$$
D_M(x) = \sqrt{(x - \mu)^T S^{-1} (x - \mu)}
$$
where:
- \( x \) is the data point.
- \( \mu \) is the mean vector.
- \( S^{-1} \) is the inverse covariance matrix.

**Description:**
- Measures the distance of a point from the mean in a **multivariate** space.
- Accounts for correlation between features.

**Interpretation:**
- **Smaller \(D_M(x)\)** → Closer to normal distribution (inliers).
- **Larger \(D_M(x)\)** → More likely an anomaly (outliers).

---

### 2. Threshold-Based Classification

**Description:**
- A threshold (\( \tau \)) is set to classify points as inliers or outliers.

**Interpretation:**
- **If** \( D_M(x) > \tau \) → Anomaly.
- **If** \( D_M(x) \leq \tau \) → Normal point.

- The **threshold** can be set using:
  - **Empirical rules** (e.g., 95% confidence region).
  - **Chi-Square Distribution**:
    $$
    \tau = \chi^2_{p, \alpha}
    $$
    where \( p \) is the number of features and \( \alpha \) is the significance level.

---

### 3. Confusion Matrix (if labeled data exists)

**Description:**
- Evaluates classification performance.

**Interpretation:**
- True Positives (TP): Correctly detected anomalies.
- False Positives (FP): Normal points wrongly classified as anomalies.
- True Negatives (TN): Correctly identified normal points.
- False Negatives (FN): Anomalies missed by the model.

---

### 4. Precision

**Formula:**
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Description:**
- Measures how many detected anomalies are actually anomalies.

**Interpretation:**
- A high precision means fewer false positives.

---

### 5. Recall

**Formula:**
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

**Description:**
- Measures the model’s ability to detect all anomalies.

**Interpretation:**
- A high recall means fewer missed anomalies.

---

### 6. F1-Score

**Formula:**
$$
F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Description:**
- A balance between precision and recall.

**Interpretation:**
- A higher F1-score means better anomaly detection.

---

### 7. Receiver Operating Characteristic (ROC) Curve & AUC

**Description:**
- ROC Curve plots **True Positive Rate (TPR)** vs. **False Positive Rate (FPR)**.
- AUC (Area Under Curve) measures model performance.

**Interpretation:**
- **AUC = 1** → Perfect classification.
- **AUC > 0.8** → Good detection ability.
- **AUC = 0.5** → Random classification.

---

### 8. Silhouette Score (for Unsupervised Evaluation)

**Formula:**
$$
S = \frac{b - a}{\max(a, b)}
$$
where:
- \( a \) is the average intra-cluster distance.
- \( b \) is the average nearest-cluster distance.

**Description:**
- Measures how well-separated anomalies are.

**Interpretation:**
- Higher values (\(\approx 1\)) indicate well-separated anomalies.

---

### 9. k-Fold Cross Validation

**Description:**
- Splits the dataset into \( k \) subsets, training the model on \( k-1 \) folds and testing on the remaining one.

**Interpretation:**
- Reduces overfitting risk.
- Provides reliable performance estimates.

---

# Mahalanobis Distance - Example

## Data loading

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from scipy.spatial.distance import mahalanobis

# Data Loading
data = load_iris()
X = data.data
y = data.target

# Use only one class (e.g., class 0) and treat the rest as outliers
X_class_0 = X[y == 0]
y_class_0 = y[y == 0]
````

##  Data processing

````python
# Data Processing
df = pd.DataFrame(X_class_0, columns=data.feature_names)
print("First few rows of the dataset:")
print(df.head())

print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_class_0)
````

## Plotting data

## Model definition

````python
# Compute Mahalanobis Distance
mean = np.mean(X_scaled, axis=0)
cov_inv = np.linalg.inv(np.cov(X_scaled.T))
mahal_distances = np.array([mahalanobis(sample, mean, cov_inv) for sample in X_scaled])

# Define threshold for outliers (e.g., 95th percentile)
threshold = np.percentile(mahal_distances, 95)
y_pred = (mahal_distances > threshold).astype(int)
````

## Model evaulation

````python

print("\nConfusion Matrix:")
print(confusion_matrix(y_class_0, y_pred))

print("\nClassification Report:")
print(classification_report(y_class_0, y_pred))

# Data Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='viridis')
plt.title('Mahalanobis Distance (First Two Features)')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
````
