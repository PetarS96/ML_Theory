# **Elliptic Envelope Model Theory**


## Elliptic Envelope

---

## Theory
Elliptic Envelope is an unsupervised anomaly detection algorithm that fits an elliptic envelope to the data. This algorithm assumes that the data is Gaussian distributed and tries to define the decision boundary using the Mahalanobis distance. Elliptic Envelope is useful for detecting multivariate outliers, especially when the data follows an elliptical distribution.

The main idea is to:
- Estimate the robust covariance of the data.
- Fit an elliptic envelope to the data points based on the Mahalanobis distance.
- Identify points outside the ellipse as outliers.

---

## Mathematical Foundation
- **Mahalanobis Distance**:
  A measure of the distance between a point \( x \) and a distribution:
  $$ D_M(x) = \sqrt{(x - \mu)^T \Sigma^{-1} (x - \mu)} $$
  where:
  - \( x \) is the data point.
  - \( \mu \) is the mean vector of the data.
  - \( \Sigma \) is the covariance matrix of the data.

- **Robust Covariance**:
  Estimates the covariance of the data using robust methods to minimize the influence of outliers:
  - **Minimum Covariance Determinant (MCD)**: Finds the subset of data points with the smallest determinant of the covariance matrix.

- **Elliptic Envelope**:
  Fits an ellipse to the data points by estimating the robust mean and covariance:
  $$ E(x) = \{ x : D_M(x) \leq \text{threshold} \} $$
  Points outside the envelope are considered outliers.

---

## Algorithm Steps
1. **Estimate Robust Covariance**:
   - Use the MCD method to estimate the robust mean and covariance matrix of the data.

2. **Fit Elliptic Envelope**:
   - Fit an ellipse to the data points based on the Mahalanobis distance.

3. **Identify Outliers**:
   - Classify points outside the elliptic envelope as outliers.

---

## Key Parameters
- **support_fraction**: The proportion of points to be included in the support of the raw MCD estimate.
- **contamination**: The proportion of anomalies in the dataset.
- **random_state**: Controls the random seed given to the initializations.

---

## Advantages
- Effective for detecting multivariate outliers in elliptical distributions.
- Provides a robust estimate of the mean and covariance.
- Works well with Gaussian distributed data.

---

## Disadvantages
- Assumes the data follows a Gaussian distribution.
- May not perform well with non-Gaussian data or data with non-elliptical distributions.
- Computationally expensive for high-dimensional data.

---

## Implementation Tips
- Use **StandardScaler** to standardize the features before fitting the model.
- Choose an appropriate **contamination** parameter based on domain knowledge or exploratory analysis.
- Combine with other anomaly detection methods for improved performance.

---

## Applications
- Fraud detection
- Quality control in manufacturing
- Outlier detection in multivariate datasets
- Financial anomaly detection

Elliptic Envelope is a powerful algorithm for detecting multivariate outliers in Gaussian distributed data. Its robust covariance estimation makes it suitable for many real-world applications where the data follows an elliptical distribution.


## Model Evaluation for Elliptic Envelope

### 1. Mahalanobis Distance

**Formula:**
$$
D_M(x) = \sqrt{(x - \mu)^T \Sigma^{-1} (x - \mu)}
$$
where:
- \( x \) is the data point.
- \( \mu \) is the mean vector of the data.
- \( \Sigma \) is the covariance matrix of the data.

**Description:**
- Measures the distance between a data point and the Gaussian distribution.
- Higher distances indicate potential outliers.

**Interpretation:**
- Points with high Mahalanobis distances are considered outliers.

---

### 2. Confusion Matrix (for labeled data)

**Description:**
- If labeled anomalies exist, a **confusion matrix** helps evaluate classification performance.

**Interpretation:**
- True Positives (TP): Correctly identified anomalies.
- False Positives (FP): Normal points misclassified as anomalies.
- True Negatives (TN): Correctly identified normal points.
- False Negatives (FN): Anomalies misclassified as normal.

---

### 3. Precision

**Formula:**
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Description:**
- Measures how many detected anomalies are actually anomalies.

**Interpretation:**
- A high precision means fewer false alarms.

---

### 4. Recall

**Formula:**
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

**Description:**
- Measures the model's ability to detect all anomalies.

**Interpretation:**
- A high recall means fewer missed anomalies.

---

### 5. F1-Score

**Formula:**
$$
F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Description:**
- A balance between precision and recall.

**Interpretation:**
- A higher F1-score means better anomaly detection performance.

---

### 6. Receiver Operating Characteristic (ROC) Curve & AUC

**Description:**
- ROC Curve plots the **True Positive Rate (TPR)** against the **False Positive Rate (FPR)**.
- AUC (Area Under Curve) measures overall classification performance.

**Interpretation:**
- **AUC = 1** → Perfect detection.
- **AUC > 0.8** → Good performance.
- **AUC = 0.5** → Random performance.

---

### 7. Adjusting Contamination Parameter

**Description:**
- The contamination parameter (\(\text{contamination}\)) defines the proportion of anomalies in the dataset.

**Interpretation:**
- A well-chosen contamination value improves model performance.
- **Too high** → More false positives.
- **Too low** → More false negatives.

---

### 8. Silhouette Score (for Unsupervised Evaluation)

**Formula:**
$$
S = \frac{b - a}{\max(a, b)}
$$
where:
- \(a\) is the average intra-cluster distance.
- \(b\) is the average nearest-cluster distance.

**Description:**
- Measures how well-separated the anomalies are.

**Interpretation:**
- Higher values (\(\approx 1\)) indicate well-separated anomalies.

---

### 9. k-Fold Cross Validation

**Description:**
- Splits the dataset into \(k\) subsets, training the model on \(k-1\) folds and testing on the remaining one.

**Interpretation:**
- Reduces overfitting risk.
- Provides more reliable performance estimates.

### sklearn.covariance.EllipticEnvelope(store_precision=True, assume_centered=False, support_fraction=None, contamination=0.1, random_state=None)

| **Parameter**   | **Description**                                                                 |
|----------------|-------------------------------------------------------------------------------|
| store_precision | If `True`, the estimated precision is stored.                                 |
| assume_centered | If `True`, the support of the robust location and covariance estimates is computed, and the robust location and covariance estimates are set to zero. |
| support_fraction | The proportion of points to be included in the support of the raw MCD estimate. |
| contamination  | The proportion of anomalies in the dataset.                                    |
| random_state   | Controls the random seed given to the initializations.                         |

-

| **Attribute**           | **Description**                                                                 |
|-------------------------|-------------------------------------------------------------------------------|
| location_               | Estimated robust location of the data.                                        |
| covariance_             | Estimated robust covariance matrix.                                           |
| precision_              | Estimated pseudo inverse matrix.                                              |
| support_                | Mask of the observations used to compute the robust estimates of location and shape. |
| offset_                 | Offset used to define the decision function.                                   |
| decision_scores_        | Raw anomaly scores for training data.                                          |

-

| **Method**              | **Description**                                                                 |
|------------------------|-------------------------------------------------------------------------------|
| fit(X[, y])            | Fit the EllipticEnvelope model on training data `X`.                           |
| predict(X)             | Predict the labels (`-1` for outliers and `1` for inliers) for data `X`.       |
| decision_function(X)   | Compute the decision function for each sample in `X`.                          |
| mahalanobis(X)         | Compute the squared Mahalanobis distances.                                      |
| set_params(**params)   | Set the parameters of the estimator.                                            |
| get_params([deep])     | Get parameters for this estimator.                                              |

-

| **Usage Example (Python)**  | **Code**                                                                 |
|----------------------------|------------------------------------------------------------------------|
| **Fit the Elliptic Envelope model** | `ee = EllipticEnvelope(contamination=0.1, random_state=42)`<br>`ee.fit(X)` |

| **Compute Mahalanobis distances** | `mahalanobis_distances = ee.mahalanobis(X)` |

## Data loading

```python
import numpy as np
import pandas as pd
from sklearn.covariance import EllipticEnvelope
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Data Loading
data = load_breast_cancer()
X = data.data
y = data.target
```

##  Data processing

```python

# Data Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Plotting data

```python

# Data Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y, palette='viridis')
plt.title('Breast Cancer Data (First Two Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

## Model definition

```python

# Model Definition and Training
ee = EllipticEnvelope(contamination=0.1, random_state=42)
ee.fit(X_scaled)

# Predict labels
y_pred = ee.predict(X_scaled)

# Convert prediction labels to binary format (1 for outliers, 0 for inliers)
y_pred_binary = np.where(y_pred == -1, 1, 0)

# Compute Mahalanobis distances
mahalanobis_distances = ee.mahalanobis(X_scaled)

# Define Threshold for Identifying Outliers
threshold = np.percentile(mahalanobis_distances, 90)
outliers = mahalanobis_distances > threshold
```

## Model evaulation

```python

# Model Evaluation
print("Confusion Matrix:")
print(confusion_matrix(outliers, y_pred_binary))

print("\nClassification Report:")
print(classification_report(outliers, y_pred_binary))

# Plotting the Results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred_binary, palette='viridis')
plt.title('Elliptic Envelope Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
