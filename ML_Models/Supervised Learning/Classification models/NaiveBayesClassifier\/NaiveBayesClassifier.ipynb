# **Naive Bayes Classifier Model Theory**


## Theory
Naive Bayes is a probabilistic classifier based on applying Bayes' theorem with the "naive" assumption of conditional independence between features. It calculates the probability of each class given the input features and selects the most probable class.
The main idea is to:
- Calculate prior probabilities from class frequencies
- Calculate likelihood of features given each class
- Apply Bayes' theorem to get posterior probabilities
- Choose class with highest posterior probability

## Mathematical Foundation
- **Bayes' Theorem**:
$$ P(y|x) = \frac{P(x|y)P(y)}{P(x)} $$

- **Naive Bayes Classifier**:
$$ \hat{y} = \argmax_{y} P(y)\prod_{i=1}^{n} P(x_i|y) $$

## Types of Naive Bayes
- **Gaussian NB**:
$$ P(x_i|y) = \frac{1}{\sqrt{2\pi\sigma^2_y}}\exp(-\frac{(x_i-\mu_y)^2}{2\sigma^2_y}) $$

- **Multinomial NB**:
$$ P(x_i|y) = \frac{\text{count}(x_i,y) + \alpha}{\text{count}(y) + \alpha n} $$

- **Bernoulli NB**:
$$ P(x_i|y) = P(i|y)x_i + (1-P(i|y))(1-x_i) $$

## Classification Process
1. **Training**:
- Calculate prior probabilities P(y)
- Calculate feature likelihoods P(x|y)
- Store parameters for each class

2. **Prediction**:
- Apply Bayes' theorem
- Choose class with highest posterior probability

## Key Parameters
- **alpha**: Smoothing parameter (Laplace/Lidstone)
- **fit_prior**: Whether to learn class prior probabilities
- **class_prior**: Prior probabilities of classes
- **var_smoothing**: Variance smoothing (Gaussian NB)

## Probability Calculations
- **Prior Probability**:
$$ P(y) = \frac{\text{count}(y)}{N} $$

- **Posterior Probability**:
$$ P(y|x_1,...,x_n) \propto P(y)\prod_{i=1}^{n} P(x_i|y) $$

## Advantages
- Simple and fast to train
- Works well with high-dimensional data
- Performs well with small training sets
- Good for text classification
- Easy to update with new training data

## Disadvantages
- Assumes feature independence
- Limited by "zero frequency" problem
- May be outperformed by more sophisticated models
- Cannot learn feature interactions
- Sensitive to irrelevant features

## Implementation Tips
- Use Laplace smoothing for zero frequencies
- Log probabilities to prevent underflow
- Feature scaling not required
- Good baseline for text classification
- Consider feature selection to remove irrelevant features

Naïve Bayes is one of the most straightforward and fast classification algorithm. It is very well suited for large volume of data. It is successfully used in various applications such as :

Spam filtering
Text classification
Sentiment analysis
Recommender systems
It uses Bayes theorem of probability for prediction of unknown class.


## Model Evaluation for Naive Bayes Classifier

### 1. Accuracy Score
Formula:
$$
\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
$$
Description:
- Measures overall prediction accuracy.
- Basic evaluation metric for balanced datasets.
Interpretation:
- Higher values indicate better overall performance.
- Should be considered alongside class-specific metrics.
---

### 2. Posterior Probability
Formula:
$$
P(C_k|x) = \frac{P(C_k)\prod_{i=1}^n P(x_i|C_k)}{P(x)}
$$
Description:
- Probability of class given the features.
- Core component of Naive Bayes prediction.
Interpretation:
- Higher values indicate stronger class prediction.
- Used for classification decisions and confidence estimation.
---

### 3. Prior Probability
Formula:
$$
P(C_k) = \frac{\text{Number of samples in class k}}{\text{Total number of samples}}
$$
Description:
- Initial probability of each class.
- Based on class distribution in training data.
Interpretation:
- Reflects class balance in training data.
- Important for understanding class bias.
---

### 4. Likelihood
Formula:
$$
P(x|C_k) = \prod_{i=1}^n P(x_i|C_k)
$$
Description:
- Probability of features given the class.
- Product of individual feature probabilities.
Interpretation:
- Measures feature support for each class.
- Key component in final prediction.
---

### 5. Log-Likelihood
Formula:
$$
\log P(x|C_k) = \sum_{i=1}^n \log P(x_i|C_k)
$$
Description:
- Logarithmic form of likelihood calculation.
- Used to prevent numerical underflow.
Interpretation:
- Higher values indicate stronger feature evidence.
- More numerically stable than raw probability.
---

### 6. Feature Importance
Description:
- Contribution of each feature to class prediction.
- Based on conditional probabilities.
Interpretation:
- Higher values indicate more discriminative features.
- Useful for feature selection and understanding.
---

### 7. Class Conditional Probabilities
Formula:
$$
P(x_i|C_k) = \frac{\text{Count}(x_i, C_k) + \alpha}{\text{Count}(C_k) + \alpha n}
$$
Description:
- Probability of feature value given class.
- Includes Laplace smoothing (α).
Interpretation:
- Handles zero frequency problem.
- Affects model's sensitivity to rare events.
---

### 8. ROC-AUC Score
Formula:
$$
\text{AUC} = \int_{0}^{1} \text{TPR}(t)\text{FPR}'(t)dt
$$
Description:
- Area under ROC curve for binary classification.
- Threshold-independent performance measure.
Interpretation:
- 1.0 indicates perfect classification.
- 0.5 represents random prediction.
---

### 9. Brier Score
Formula:
$$
\text{BS} = \frac{1}{N}\sum_{i=1}^N (f_i - o_i)^2
$$
Description:
- Measures accuracy of probabilistic predictions.
- Squared error of predicted probabilities.
Interpretation:
- Lower values indicate better calibrated probabilities.
- Useful for assessing probability estimates.
---

### 10. Cross-Entropy Loss
Formula:
$$
H(p,q) = -\sum_{i} p(x_i)\log q(x_i)
$$
Description:
- Measures discrepancy between true and predicted probabilities.
- Common loss function for probabilistic classifiers.
Interpretation:
- Lower values indicate better probability estimates.
- Sensitive to probability calibration.
---

## sklearn template [GaussianNB](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html)

### class sklearn.naive_bayes.GaussianNB(*, priors=None, var_smoothing=1e-9)

| **Parameter**               | **Description**                                                                                                                                     | **Default**      |
|----------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------|------------------|
| `priors`                   | Prior probabilities of the classes. If not given, the classes are assumed to have equal prior probabilities                                          | `None`           |
| `var_smoothing`            | Portion of the largest variance of all features that is added to variances for calculation stability                                                 | `1e-9`           |

-

| **Attribute**              | **Description**                                                                                                                                     |
|----------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------|
| `classes_`                 | The class labels                                                                                                                                    |
| `class_prior_`             | Probability of each class                                                                                                                           |
| `class_count_`             | Number of training samples observed in each class                                                                                                   |
| `theta_`                   | Mean of each feature per class                                                                                                                      |
| `var_`                     | Variance of each feature per class                                                                                                                  |
| `epsilon_`                 | Absolute additive value to variances                                                                                                                |
| `n_features_in_`           | Number of features seen during fit                                                                                                                  |

-

| **Method**                 | **Description**                                                                                                                                     |
|----------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------|
| `fit(X, y)`                | Fit Gaussian Naive Bayes according to X, y                                                                                                          |
| `predict(X)`               | Perform classification on test vectors X                                                                                                            |
| `predict_proba(X)`         | Return probability estimates for the test vectors X                                                                                                 |
| `predict_log_proba(X)`     | Return log-probability estimates for the test vectors X                                                                                             |
| `score(X, y)`              | Return the mean accuracy on the given test data and labels                                                                                          |
| `partial_fit(X, y)`        | Incremental fit on a batch of samples                                                                                                               |

-

#### Other Naive Bayes Variants:
- `MultinomialNB`: For discrete features (e.g., word counts for text classification)
- `BernoulliNB`: For binary/boolean features
- `ComplementNB`: For text classification with imbalanced classes
- `CategoricalNB`: For categorical features

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

data = '/home/petar-ubuntu/Learning/ML_Theory/ML_Models/Supervised Learning/Classification models/NaiveBayesClassifier\/data/adult.csv'

df = pd.read_csv(data, header=None, sep=',\s')

col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

df.head()
```

## Plotting data

```python
import matplotlib.pyplot as plt # for data visualization purposes
import seaborn as sns # for statistical data visualization
```

## Model definition

```python
X = df.drop(['income'], axis=1)

y = df['income']

# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# find numerical variables

numerical = [var for var in df.columns if df[var].dtype!='O']

print('There are {} numerical variables\n'.format(len(numerical)))

print('The numerical variables are :', numerical)
```

```python
# display categorical variables

categorical = [col for col in X_train.columns if X_train[col].dtypes == 'O']

categorical
```

```python
for df2 in [X_train, X_test]:
    df2['workclass'].fillna(X_train['workclass'].mode()[0], inplace=True)
    df2['occupation'].fillna(X_train['occupation'].mode()[0], inplace=True)
    df2['native_country'].fillna(X_train['native_country'].mode()[0], inplace=True)    
```

```python
# import category encoders

import category_encoders as ce

# encode remaining variables with one-hot encoding

encoder = ce.OneHotEncoder(cols=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 
                                 'race', 'sex', 'native_country'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
cols = X_train.columns

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns=[cols])

X_test = pd.DataFrame(X_test, columns=[cols])
```

## Model evaulation

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = gnb.predict(X_train)

y_pred_train

# print the scores on training and test set

print('Training set score: {:.4f}'.format(gnb.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(gnb.score(X_test, y_test)))
```

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])

# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]

# print classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))

# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))

# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))

recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))

true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))

false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))

specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```

```python
# store the predicted probabilities for class 1 - Probability of >50K

y_pred1 = gnb.predict_proba(X_test)[:, 1]

# plot histogram of predicted probabilities


# adjust the font size 
plt.rcParams['font.size'] = 12


# plot histogram with 10 bins
plt.hist(y_pred1, bins = 10)


# set the title of predicted probabilities
plt.title('Histogram of predicted probabilities of salaries >50K')


# set the x-axis limit
plt.xlim(0,1)


# set the title
plt.xlabel('Predicted probabilities of salaries >50K')
plt.ylabel('Frequency')
```

```python
# plot ROC Curve

from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred1, pos_label = '>50K')

plt.figure(figsize=(6,4))

plt.plot(fpr, tpr, linewidth=2)

plt.plot([0,1], [0,1], 'k--' )

plt.rcParams['font.size'] = 12

plt.title('ROC curve for Gaussian Naive Bayes Classifier for Predicting Salaries')

plt.xlabel('False Positive Rate (1 - Specificity)')

plt.ylabel('True Positive Rate (Sensitivity)')

plt.show()
```

```python
# compute ROC AUC

from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred1)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

```python
# calculate cross-validated ROC AUC 

from sklearn.model_selection import cross_val_score

Cross_validated_ROC_AUC = cross_val_score(gnb, X_train, y_train, cv=5, scoring='roc_auc').mean()

print('Cross validated ROC AUC : {:.4f}'.format(Cross_validated_ROC_AUC))
```

```python
# Applying 10-Fold Cross Validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(gnb, X_train, y_train, cv = 10, scoring='accuracy')

print('Cross-validation scores:{}'.format(scores))

# compute Average cross-validation score

print('Average cross-validation score: {:.4f}'.format(scores.mean()))
```
