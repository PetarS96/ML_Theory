# **Gaussian Mixture Model Model Theory**


## Gaussian Mixture Model (GMM)

---

## Theory
Gaussian Mixture Model (GMM) is a probabilistic model used to represent a mixture of multiple Gaussian distributions within a dataset. It is widely used for clustering, density estimation, and anomaly detection. GMM assumes that the data is generated from a mixture of several Gaussian distributions, each characterized by its mean and covariance. The algorithm aims to identify the parameters of these Gaussian components using the Expectation-Maximization (EM) algorithm.

The main idea is to:
- Model the data as a combination of multiple Gaussian distributions.
- Use the EM algorithm to iteratively estimate the parameters of these distributions.
- Assign probabilities to data points based on their likelihood of belonging to each Gaussian component.

---

## Mathematical Foundation
- **Gaussian Mixture**:
  A weighted sum of multiple Gaussian distributions:
  $$ p(x) = \sum_{k=1}^K \pi_k \mathcal{N}(x | \mu_k, \Sigma_k) $$
  - \( \pi_k \): Mixture weight for the \( k \)-th component.
  - \( \mathcal{N}(x | \mu_k, \Sigma_k) \): Gaussian distribution with mean \( \mu_k \) and covariance \( \Sigma_k \).
  - \( K \): Number of components.

- **Expectation-Maximization (EM) Algorithm**:
  An iterative method to estimate the parameters of the Gaussian components:
  1. **Expectation (E) Step**:
     - Calculate the responsibility \( \gamma_{ik} \) for each data point \( x_i \) and component \( k \):
       $$ \gamma_{ik} = \frac{\pi_k \mathcal{N}(x_i | \mu_k, \Sigma_k)}{\sum_{j=1}^K \pi_j \mathcal{N}(x_i | \mu_j, \Sigma_j)} $$
  2. **Maximization (M) Step**:
     - Update the parameters based on the responsibilities:
       $$ \mu_k = \frac{\sum_{i=1}^n \gamma_{ik} x_i}{\sum_{i=1}^n \gamma_{ik}} $$
       $$ \Sigma_k = \frac{\sum_{i=1}^n \gamma_{ik} (x_i - \mu_k)(x_i - \mu_k)^T}{\sum_{i=1}^n \gamma_{ik}} $$
       $$ \pi_k = \frac{\sum_{i=1}^n \gamma_{ik}}{n} $$

- **Log-Likelihood**:
  Measures the probability of the observed data given the model parameters:
  $$ \text{Log-Likelihood} = \sum_{i=1}^n \log \left( \sum_{k=1}^K \pi_k \mathcal{N}(x_i | \mu_k, \Sigma_k) \right) $$

---

## Algorithm Steps
1. **Initialization**:
   - Initialize the parameters of the Gaussian components (\( \pi_k, \mu_k, \Sigma_k \)).

2. **E-Step**:
   - Calculate the responsibilities for each data point.

3. **M-Step**:
   - Update the parameters based on the responsibilities.

4. **Convergence**:
   - Repeat the E and M steps until convergence (i.e., the log-likelihood changes below a threshold).

---

## Key Parameters
- **n_components**: The number of Gaussian components.
- **covariance_type**: The type of covariance parameters (`full`, `tied`, `diag`, `spherical`).
- **max_iter**: The maximum number of iterations for the EM algorithm.
- **random_state**: Controls the random seed given to the initializations.

---

## Advantages
- Flexible in modeling complex distributions.
- Provides probabilistic cluster assignments.
- Can handle varying shapes and sizes of clusters.

---

## Disadvantages
- Sensitive to the choice of the number of components.
- May converge to local optima.
- Requires careful initialization of parameters.

---

## Implementation Tips
- Use **cross-validation** or **information criteria (AIC, BIC)** to determine the optimal number of components.
- Initialize parameters using methods like **k-means** or **random initialization**.
- Regularize covariance matrices to avoid numerical instability.

---

## Applications
- Clustering
- Density estimation
- Anomaly detection
- Image segmentation
- Speech recognition

Gaussian Mixture Model is a powerful and flexible algorithm for various applications. It provides a probabilistic framework for clustering and density estimation, making it a valuable tool in many fields.


## **Model Evaluation for Gaussian Mixture Model (GMM)**

### 1. Log-Likelihood

**Formula:**
$$
\text{Log-Likelihood} = \sum_{i=1}^n \log \left( \sum_{k=1}^K \pi_k \mathcal{N}(x_i | \mu_k, \Sigma_k) \right)
$$
where:
- \(x_i\) is the \(i\)-th data point.
- \(\pi_k\) is the mixture weight for the \(k\)-th component.
- \(\mathcal{N}\) is the Gaussian distribution with mean \(\mu_k\) and covariance \(\Sigma_k\).
- \(n\) is the number of samples.
- \(K\) is the number of components.

**Description:**
- Measures the probability of the observed data given the model parameters.
- Higher values indicate a better fit.

**Interpretation:**
- Higher log-likelihood values indicate that the model explains the data well.

---

### 2. Akaike Information Criterion (AIC)

**Formula:**
$$
\text{AIC} = 2k - 2\ln(L)
$$
where:
- \(k\) is the number of parameters in the model.
- \(L\) is the likelihood of the model.

**Description:**
- Penalizes the complexity of the model to prevent overfitting.
- Lower values indicate a better model.

**Interpretation:**
- A lower AIC value suggests a better trade-off between model complexity and goodness of fit.

---

### 3. Bayesian Information Criterion (BIC)

**Formula:**
$$
\text{BIC} = \ln(n)k - 2\ln(L)
$$
where:
- \(n\) is the number of samples.
- \(k\) is the number of parameters in the model.
- \(L\) is the likelihood of the model.

**Description:**
- Similar to AIC but with a stronger penalty for the number of parameters.
- Lower values indicate a better model.

**Interpretation:**
- A lower BIC value suggests a better model with a good balance between complexity and fit.

---

### 4. Confusion Matrix (for labeled data)

**Description:**
- If labeled clusters exist, a **confusion matrix** helps evaluate classification performance.

**Interpretation:**
- True Positives (TP): Correctly identified data points in the true cluster.
- False Positives (FP): Data points incorrectly assigned to the cluster.
- True Negatives (TN): Correctly identified data points not in the cluster.
- False Negatives (FN): Data points incorrectly assigned to other clusters.

---

### 5. Precision

**Formula:**
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Description:**
- Measures how many detected data points are correctly clustered.

**Interpretation:**
- A high precision means fewer false positives.

---

### 6. Recall

**Formula:**
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

**Description:**
- Measures the model's ability to correctly identify all data points in a cluster.

**Interpretation:**
- A high recall means fewer missed points in the cluster.

---

### 7. F1-Score

**Formula:**
$$
F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Description:**
- A balance between precision and recall.

**Interpretation:**
- A higher F1-score means better clustering performance.

---

### 8. Silhouette Score

**Formula:**
$$
S = \frac{b - a}{\max(a, b)}
$$
where:
- \(a\) is the average intra-cluster distance.
- \(b\) is the average nearest-cluster distance.

**Description:**
- Measures how similar data points are within a cluster compared to other clusters.

**Interpretation:**
- Higher values (\(\approx 1\)) indicate well-separated and cohesive clusters.

---

### 9. Adjusted Rand Index (ARI)

**Formula:**
$$
\text{ARI} = \frac{\text{RI} - \text{Expected RI}}{\text{Max RI} - \text{Expected RI}}
$$
where:
- RI is the Rand Index.

**Description:**
- Measures the similarity between the true cluster assignments and the ones predicted by the model.

**Interpretation:**
- Higher ARI values indicate better clustering performance.

---

### 10. Cross-Validation

**Description:**
- Splits the dataset into multiple subsets, training the model on some subsets and testing on the remaining ones.

**Interpretation:**
- Provides a more reliable estimate of model performance and helps prevent overfitting.

---

### 11. Visualization Quality

**Description:**
- Evaluates the visual interpretability of the GMM clustering results.
- Subjective measure based on the clarity of identified clusters.

**Interpretation:**
- Clear and well-separated clusters indicate high-quality visualization.
- Useful for exploratory data analysis.

---

### 12. Computational Efficiency

**Description:**
- Evaluates the computational cost of fitting the GMM, including time and memory usage.
- Important for large datasets or real-time applications.

**Interpretation:**
- Lower computational cost indicates better scalability.
- Useful for assessing the practicality of GMM for specific use cases.


## Gaussian Mixture Model (GMM)

### sklearn.mixture.GaussianMixture(n_components=1, covariance_type='full', max_iter=100, random_state=None)

| **Parameter**   | **Description**                                                                 |
|---------------|-------------------------------------------------------------------------------|
| n_components  | Number of mixture components.                                                 |
| covariance_type | Type of covariance parameters (`full`, `tied`, `diag`, `spherical`).         |
| max_iter      | Maximum number of iterations for the EM algorithm.                            |
| random_state  | Controls the random seed given to the initializations.                       |

-

| **Attribute**           | **Description**                                                                 |
|-------------------------|-------------------------------------------------------------------------------|
| weights_                | The weights of each mixture component.                                        |
| means_                  | The mean of each mixture component.                                           |
| covariances_            | The covariance of each mixture component.                                     |
| converged_              | Whether the EM algorithm has converged.                                       |

-

| **Method**             | **Description**                                                                 |
|------------------------|-------------------------------------------------------------------------------|
| fit(X[, y])            | Fit the GMM model on training data `X`.                                        |
| predict(X)             | Predict the labels for the data samples in `X` using trained model.            |
| predict_proba(X)       | Predict the posterior probabilities of each component for each sample.         |
| score_samples(X)       | Compute the log-likelihood of each sample.                                     |
| sample([n_samples])    | Generate random samples from the fitted GMM.                                   |

-

| **Usage Example (Python)**  | **Code**                                                                 |
|----------------------------|------------------------------------------------------------------------|
| **Fit the GMM model** | `gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=42)`<br> `gmm.fit(X)` |
| **Predict labels** | `labels = gmm.predict(X)` |
| **Compute log-likelihood** | `log_likelihood = gmm.score_samples(X)` |

# GaussianMixture - Example

## Data loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.metrics import confusion_matrix, classification_report

# Data Loading
data = load_iris()
X = data.data
y = data.target

# Use only one class (e.g., class 0) and treat the rest as outliers
X_class_0 = X[y == 0]
y_class_0 = y[y == 0]

```

##  Data processing

```python

# Data Processing
df = pd.DataFrame(X_class_0, columns=data.feature_names)
print("First few rows of the dataset:")
print(df.head())

print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_class_0)
```

## Plotting data

## Model definition

```python

# Fit Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=1)
gmm.fit(X_scaled)
scores = gmm.score_samples(X_scaled)

# Define threshold for outliers (e.g., 95th percentile)
threshold = np.percentile(scores, 5)
y_pred = (scores < threshold).astype(int)
```

## Model evaulation

```python

print("\nConfusion Matrix:")
print(confusion_matrix(y_class_0, y_pred))

print("\nClassification Report:")
print(classification_report(y_class_0, y_pred))

# Data Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='viridis')
plt.title('Gaussian Mixture Model (First Two Features)')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
```
