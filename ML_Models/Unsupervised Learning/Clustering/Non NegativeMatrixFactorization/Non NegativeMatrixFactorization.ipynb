# **Non Negative Matrix Factorization Model Theory**


## Non-Negative Matrix Factorization (NMF)

---

## Theory
Non-Negative Matrix Factorization (NMF) is a dimensionality reduction technique that decomposes a non-negative matrix \( V \) into the product of two non-negative matrices \( W \) and \( H \). The goal is to approximate the original matrix \( V \) as:
\[ V \approx WH \]
where \( W \) and \( H \) are non-negative matrices. NMF is particularly useful for data that can be represented as non-negative values, such as images, text, and audio data. It is widely used in applications like topic modeling, image analysis, and recommendation systems.

The main idea is to:
- Decompose the non-negative data matrix into two non-negative matrices.
- Ensure that the factorized matrices \( W \) and \( H \) are non-negative.
- Minimize the reconstruction error between the original matrix and the product of the factorized matrices.

---

## Mathematical Foundation
- **Matrix Decomposition**:
  Given a non-negative matrix \( V \) of size \( m \times n \), NMF decomposes it into two non-negative matrices \( W \) and \( H \):
  \[ V \approx WH \]
  - \( V \): Original data matrix.
  - \( W \): Basis matrix of size \( m \times k \).
  - \( H \): Coefficient matrix of size \( k \times n \).
  - \( k \): Number of components (or factors).

- **Reconstruction Error**:
  The objective is to minimize the reconstruction error, typically measured using the Frobenius norm or Kullback-Leibler divergence:
  \[ \min_{W, H} \| V - WH \|_F^2 \]
  or
  \[ \min_{W, H} D_{\text{KL}}(V \| WH) \]
  - \( \| \cdot \|_F \): Frobenius norm.
  - \( D_{\text{KL}} \): Kullback-Leibler divergence.

- **Non-Negativity Constraint**:
  Both \( W \) and \( H \) must be non-negative:
  \[ W \geq 0, \quad H \geq 0 \]

---

## Algorithm Steps
1. **Initialize Matrices**:
   - Initialize the basis matrix \( W \) and the coefficient matrix \( H \) with non-negative values. Common initialization methods include random initialization and non-negative double singular value decomposition (NNDSVD).

2. **Update Rules**:
   - Update \( W \) and \( H \) iteratively to minimize the reconstruction error. Common update rules include multiplicative update rules and gradient descent methods.

3. **Multiplicative Update Rules**:
   - Update \( W \) and \( H \) using the following multiplicative rules:
     \[ W_{ij} \leftarrow W_{ij} \cdot \frac{(VH^T)_{ij}}{(WHH^T)_{ij}} \]
     \[ H_{ij} \leftarrow H_{ij} \cdot \frac{(W^TV)_{ij}}{(W^TWH)_{ij}} \]

4. **Convergence Check**:
   - Check for convergence by monitoring the change in the reconstruction error or the matrices \( W \) and \( H \).

5. **Repeat**:
   - Repeat the update and convergence check steps until the reconstruction error is minimized or a maximum number of iterations is reached.

---

## Key Parameters
- **n_components**: The number of components (or factors) to extract.
- **init**: The method used to initialize the matrices \( W \) and \( H \) (e.g., `random`, `nndsvd`).
- **solver**: The optimization algorithm used to update \( W \) and \( H \) (e.g., `mu`, `cd`).
- **beta_loss**: The type of loss function to minimize (e.g., `frobenius`, `kullback-leibler`).
- **max_iter**: The maximum number of iterations for the optimization process.
- **tol**: The tolerance for stopping the optimization process.

---

## Advantages
- Ensures non-negative components, which can be more interpretable in many applications.
- Useful for data with natural non-negative constraints, such as images, text, and audio.
- Can handle sparse data and missing values.
- Provides a parts-based representation of the data, which can be useful for feature extraction and clustering.

---

## Disadvantages
- Computationally expensive, especially for large matrices and high-dimensional data.
- The solution is not unique and can depend on the initialization.
- May not capture global structure as well as other methods like PCA.
- The number of components to retain is often subjective and can be difficult to determine.

---

## Implementation Tips
- **Standardize** the data if necessary, but ensure that the data remains non-negative.
- Use **NNDSVD** initialization for better convergence and more stable results.
- Experiment with different values of **n_components** to find the optimal number of factors.
- Use **multiplicative update rules** for simplicity and stability, or **coordinate descent** for faster convergence.
- Validate the results using additional data or domain knowledge.

---

## Applications
- **Topic Modeling**: Extracting topics from text data.
- **Image Analysis**: Decomposing images into parts-based representations.
- **Audio Analysis**: Analyzing and separating audio signals.
- **Recommendation Systems**: Recommending items based on user preferences.
- **Biology and Genomics**: Analyzing gene expression data and identifying gene clusters.
- **Market Basket Analysis**: Identifying patterns in customer purchase data.

Non-Negative Matrix Factorization is a powerful tool for dimensionality reduction and feature extraction in non-negative data. While it has some limitations, it is particularly useful for applications where non-negative constraints are natural and interpretable.

# Model Evaluation for Non-negative Matrix Factorization (NMF)

---

### 1. Reconstruction Error
**Formula:**
$$
\text{Reconstruction Error} = \frac{1}{N} \sum_{i=1}^N ||X - \hat{X}||^2
$$
**Description:**
- Measures the difference between the original data matrix \( X \) and the reconstructed data matrix \( \hat{X} \).
- Commonly used loss function in NMF.

**Interpretation:**
- Lower values indicate better reconstruction.
- Useful for assessing the quality of the factorization.

---

### 2. Frobenius Norm
**Formula:**
$$
\text{Frobenius Norm} = ||X - \hat{X}||_F
$$
**Description:**
- Measures the Euclidean distance between the original and reconstructed matrices.
- Commonly used in NMF to minimize the reconstruction error.

**Interpretation:**
- Lower values indicate better reconstruction.
- Useful for assessing the quality of the factorization.

---

### 3. Explained Variance
**Formula:**
$$
\text{Explained Variance} = 1 - \frac{||X - \hat{X}||_F^2}{||X||_F^2}
$$
**Description:**
- Measures the proportion of the total variance in the data that is explained by the factorization.
- \( ||X||_F \) is the Frobenius norm of the original data matrix \( X \).

**Interpretation:**
- Higher values indicate that the factors capture more variance.
- Useful for assessing the effectiveness of the factorization.

---

### 4. Sparsity of Factors
**Description:**
- Measures the sparsity of the factor matrices \( W \) and \( H \).
- Sparsity can be quantified using the L1 norm or the proportion of zero elements.

**Interpretation:**
- Higher sparsity indicates that the factors are more interpretable.
- Useful for understanding the structure of the factors.

---

### 5. Feature Visualization
**Description:**
- Visualizes the learned features in the factor matrices \( W \) and \( H \).
- Helps understand what the NMF has learned.

**Interpretation:**
- Clear and meaningful features indicate good learning.
- Useful for interpretability and debugging.

---

### 6. Latent Space Analysis
**Description:**
- Evaluates the structure and properties of the latent space represented by the factor matrix \( H \).
- Can include visualizations like t-SNE or PCA of the latent space.

**Interpretation:**
- Well-structured latent space indicates effective dimensionality reduction.
- Useful for understanding the NMF's internal representation.

---

### 7. Reconstruction Quality on Test Data
**Description:**
- Evaluates the NMF's ability to generalize to unseen data.
- Measures reconstruction error on a test set.

**Interpretation:**
- Lower reconstruction error on test data indicates better generalization.
- Useful for assessing overfitting.

---

### 8. Dimensionality Reduction Effectiveness
**Description:**
- Evaluates the effectiveness of NMF in reducing the number of dimensions while retaining information.
- Measured by comparing performance metrics (e.g., classification accuracy) before and after dimensionality reduction.

**Interpretation:**
- Higher retained performance indicates effective dimensionality reduction.
- Useful for assessing the trade-off between complexity and information loss.

---

### 9. Computational Efficiency
**Description:**
- Evaluates the computational cost of training and using NMF, including time and memory usage.
- Important for large datasets or real-time applications.

**Interpretation:**
- Lower computational cost indicates better scalability.
- Useful for assessing the practicality of NMF for specific use cases.

---

### 10. Robustness to Noise
**Description:**
- Evaluates the NMF's ability to handle noisy input data.
- Measures reconstruction error on noisy data.

**Interpretation:**
- Lower reconstruction error on noisy data indicates better robustness.
- Useful for assessing the NMF's ability to denoise.

---

### 11. Anomaly Detection
**Description:**
- Evaluates the NMF's ability to detect outliers by examining the reconstruction error.
- Points with high reconstruction error are potential anomalies.

**Interpretation:**
- Points with high reconstruction error are likely outliers.
- Useful for identifying anomalies in the data.

---

### 12. Overfitting and Underfitting
**Description:**
- Evaluates the NMF's performance on training and validation sets.

- Underfitting occurs when the model performs poorly on both sets.

**Interpretation:**

# Non Negative Matrix Factorization - Example

## Data loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.decomposition import NMF
from sklearn.preprocessing import StandardScaler

# Data Loading
# Load the Digits dataset from sklearn.datasets
data = load_digits()
X = data.data
y = data.target

# Data Processing
# Convert the data to a DataFrame for better visualization
df = pd.DataFrame(data.data, columns=[f'pixel{i}' for i in range(64)])
df['target'] = data.target

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

##  Data processing

```python
# Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features (NMF works with non-negative data, so we don't standardize but ensure non-negativity)
X_non_negative = np.maximum(X, 0)  # Ensure all values are non-negative
```

## Plotting data

```python
# Data Plotting
# Plot a few sample images from the dataset
n_samples_to_plot = 5
fig, axes = plt.subplots(1, n_samples_to_plot, figsize=(15, 3))
for i in range(n_samples_to_plot):
    axes[i].imshow(X_non_negative[i].reshape(8, 8), cmap='gray')
    axes[i].set_title(f"Label: {y[i]}")
    axes[i].axis('off')
plt.show()
```

## Model definition

```python
# Model Definition
# Define the NMF model
n_components = 10  # Number of components to extract
nmf = NMF(n_components=n_components, random_state=42)

# Model Training
# Fit the NMF model to the non-negative data
W = nmf.fit_transform(X_non_negative)
H = nmf.components_

# Convert the transformed data to a DataFrame
df_nmf = pd.DataFrame(W, columns=[f'Component{i+1}' for i in range(n_components)])
df_nmf['target'] = y
```

## Model evaulation

```python
# Data Plotting
# Plot the transformed data to visualize the NMF components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Component1', y='Component2', hue='target', data=df_nmf, palette='viridis')
plt.title('NMF Components (Component1 vs Component2)')
plt.show()

# Plot the basis vectors (components)
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(H[i].reshape(8, 8), cmap='gray')
    ax.set_title(f'Component {i+1}')
    ax.axis('off')
plt.show()

# Additional Analysis
# Print the reconstruction error
print("\nReconstruction Error:")
print(nmf.reconstruction_err_)

# Print the components (basis vectors)
print("\nComponents (Basis Vectors):")
print(H)
```
