# **Connectivity-Based Outlier Factor (COF) Model Theory**


## Connectivity-Based Outlier Factor (COF)

---

## Theory
Connectivity-Based Outlier Factor (COF) is an unsupervised anomaly detection algorithm that identifies outliers based on **local connectivity** rather than direct distance or density. It extends the idea of **Local Outlier Factor (LOF)** by considering the **connectivity paths** between points, making it more effective in datasets with varying densities.

The main idea is to:
- Compute local connectivity by analyzing paths between k-nearest neighbors.
- Compare the connectivity of each point with its neighbors.
- Assign an **outlier score** based on how disconnected a point is from its surroundings.

---

## Mathematical Foundation
- **k-Nearest Neighbors (k-NN)**:
  - For each point \( x \), find its **k-nearest neighbors** \( N_k(x) \).

- **Mean k-Neighbor Distance**:
  - Compute the average distance to k-nearest neighbors:
    $$ d_k(x) = \frac{1}{k} \sum_{i=1}^{k} d(x, x_i) $$

- **Connectivity Path Distance (CPD)**:
  - Define the local connectivity path distance:
    $$ CPD_k(x) = \frac{1}{k} \sum_{i=1}^{k} CPD_k(x_i) $$

- **Connectivity-Based Outlier Factor (COF)**:
  - The COF score is computed as:
    $$ COF_k(x) = \frac{CPD_k(x)}{d_k(x)} $$
  - **Interpretation**:
    - \( COF_k(x) \approx 1 \) → Well-connected (normal point).
    - \( COF_k(x) > 1 \) → Weakly connected (potential outlier).
    - The higher the COF, the more **outlier-like** the point.

---

## Algorithm Steps
1. **Find k-Nearest Neighbors**:
   - Identify the k-nearest neighbors for each data point.

2. **Compute Mean k-Neighbor Distance**:
   - Calculate \( d_k(x) \) for all points.

3. **Compute Connectivity Path Distance (CPD)**:
   - Compute the local connectivity distance using nearest neighbors.

4. **Calculate COF Score**:
   - Compute \( COF_k(x) \) for each data point.

5. **Classify Outliers**:
   - Points with **high COF scores** are outliers.

---

## Key Parameters
- **k**: Number of nearest neighbors to consider.
- **Distance Metric**: Euclidean, Manhattan, or Mahalanobis distance.
- **Threshold**: A COF score cutoff to classify outliers.

---

## Advantages
- **Works in Varying Density Regions**: More effective than distance-based methods.
- **Better Local Outlier Detection**: Considers neighborhood connectivity.
- **Handles Different Data Distributions**: Works well in high-dimensional datasets.

---

## Disadvantages
- **Computationally Expensive**: Requires k-NN search for all points.
- **Sensitive to k**: Choosing the right k is crucial for accuracy.
- **Harder to Interpret**: COF values do not have a fixed scale.

---

## Implementation Tips
- **Optimize k Selection**: Use domain knowledge or empirical tuning.
- **Choose an Efficient k-NN Algorithm**: KD-Tree or Ball-Tree for large datasets.
- **Scale Features**: Normalize data to improve distance calculations.
- **Combine with Other Methods**: Use alongside LOF or Isolation Forest for better detection.

---

## Applications
- **Fraud Detection**: Identifying fraudulent transactions.
- **Network Security**: Detecting unusual patterns in network traffic.
- **Medical Anomaly Detection**: Finding rare medical conditions.
- **Industrial Monitoring**: Identifying failures in manufacturing systems.

COF is a powerful anomaly detection method that extends LOF by focusing on **connectivity paths** instead of simple density. It is particularly useful in **complex, high-dimensional datasets** where traditional methods struggle.


# Model Evaluation for Connectivity-Based Outlier Factor

---

### 1. Outlier Score
**Formula:**
$$
\text{Outlier Score} = \frac{\sum_{i \neq j}^k \text{Connectivity}(i, j)}{k}
$$
**Description:**
- Measures the connectivity of each point with its neighbors.
- Higher values indicate a higher likelihood of being an outlier.

**Interpretation:**
- Lower values indicate points that are more similar to their neighbors.
- Useful for identifying potential outliers.

---

### 2. Local Outlier Factor (LOF)
**Formula:**
$$
\text{LOF}(i) = \frac{1}{k} \sum_{j \in N_k(i)} \frac{\text{Local Reachability Density}(j)}{\text{Local Reachability Density}(i)}
$$
**Description:**
- Measures the local density deviation of a point with respect to its neighbors.
- Points with higher LOF values are considered outliers.

**Interpretation:**
- LOF values > 1 indicate potential outliers.
- Useful for detecting local outliers.

---

### 3. Density-Based Spatial Clustering of Applications with Noise (DBSCAN)
**Formula:**
$$
\text{Core Distance}(i) = \epsilon, \quad \text{if } |N_\epsilon(i)| \geq \text{MinPts}
$$
**Description:**
- Identifies core points, border points, and noise.
- Noise points are considered outliers.

**Interpretation:**
- Points that are not within the neighborhood of any core point are considered outliers.
- Useful for clustering and outlier detection.

---

### 4. k-Nearest Neighbors Distance
**Formula:**
$$
\text{k-NN Distance}(i) = \frac{1}{k} \sum_{j \in N_k(i)} \text{Distance}(i, j)
$$
**Description:**
- Measures the average distance of a point to its k-nearest neighbors.
- Higher values indicate a higher likelihood of being an outlier.

**Interpretation:**
- Points with significantly higher k-NN distances compared to others are potential outliers.
- Useful for distance-based outlier detection.

---

### 5. Isolation Forest
**Formula:**
$$
\text{Anomaly Score}(i) = 2^{-\frac{\text{Path Length}(i)}{c(n)}}
$$
**Description:**
- Uses random forests to isolate points.
- Points that are isolated quickly are considered outliers.

**Interpretation:**
- Higher anomaly scores indicate potential outliers.
- Useful for detecting outliers in high-dimensional spaces.

---

### 6. Connectivity-Based Outlier Factor (COF)
**Formula:**
$$
\text{COF}(i) = \frac{\sum_{j \in N_k(i)} \text{Shortest Path Length}(i, j)}{k}
$$
**Description:**
- Measures the average shortest path length of a point to its k-nearest neighbors.
- Higher values indicate a higher likelihood of being an outlier.

**Interpretation:**
- Lower COF values indicate points that are well-connected to their neighbors.
- Useful for detecting outliers based on connectivity.

---

### 7. Cluster-Based Local Outlier Factor (CBLOF)
**Formula:**
$$
\text{CBLOF}(i) = \text{Cluster Size}(C_i) \times \text{Distance}(i, C_i)
$$
**Description:**
- Combines cluster size and distance to identify outliers.
- Points far from large clusters are considered outliers.

**Interpretation:**
- Higher CBLOF values indicate potential outliers.
- Useful for combining clustering and outlier detection.

---

### 8. Ensemble-Based Outlier Detection
**Description:**
- Combines multiple outlier detection methods to improve robustness.
- Useful for comprehensive outlier detection.

**Interpretation:**
- Consensus among methods increases confidence in outlier detection.
- Useful for reducing false positives.

# Compute Connectivity-Based Outlier Factor  - Example

## Data loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import NearestNeighbors

# Data Loading
data = load_iris()
X = data.data
y = data.target

# Use only one class (e.g., class 0) and treat the rest as outliers
X_class_0 = X[y == 0]
y_class_0 = y[y == 0]

```

##  Data processing

```python

# Data Processing
df = pd.DataFrame(X_class_0, columns=data.feature_names)
print("First few rows of the dataset:")
print(df.head())

print("\nChecking for missing values:")
print(df.isnull().sum())

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_class_0)

```

## Plotting data

## Model definition

```python
# Compute Connectivity-Based Outlier Factor (COF)
k = 5
neighbors = NearestNeighbors(n_neighbors=k)
neighbors.fit(X_scaled)
distances, indices = neighbors.kneighbors(X_scaled)

# Compute the COF
cof_scores = []
for i in range(len(X_scaled)):
    path_lengths = []
    for j in indices[i]:
        if i != j:
            path_lengths.append(np.linalg.norm(X_scaled[i] - X_scaled[j]))
    cof_scores.append(np.mean(path_lengths))

```

## Model evaulation

```python
# Define threshold for outliers (e.g., 95th percentile)
threshold = np.percentile(cof_scores, 95)
y_pred = (np.array(cof_scores) > threshold).astype(int)

print("\nConfusion Matrix:")
print(confusion_matrix(y_class_0, y_pred))

print("\nClassification Report:")
print(classification_report(y_class_0, y_pred))

# Data Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='viridis')
plt.title('Connectivity-Based Outlier Factor (First Two Features)')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
```
